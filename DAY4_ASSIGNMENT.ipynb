3. Pie chart of Indian Car companies market share in percentage(max 6 campanies)

```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px

```

**4. Covid-19 Active Cases**


```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Covid19July72020.csv")
df
```

```python
df.max()
```

```python
Active_Cases= df[["State/UnionTerritory"," Active Cases"]]
Active_Cases
```

```python
Maximum_ActiveCases = Active_Cases.nlargest(11,' Active Cases')
Maximum_ActiveCases
```

**Bar Chart(Maximum 10 Active cases)**

```python
fig = px.bar(Maximum_ActiveCases,x=' Active Cases',y='State/UnionTerritory')
fig.show()
```
